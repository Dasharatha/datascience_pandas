```python
# memory usage: 2.7 MB
df['age'] = df['age'].astype('int8')
df['income'] = df['income'].astype('float16')
df['visits'] = df['visits'].astype('int8')
df['nvisits'] = df['nvisits'].astype('int8')
df['ovisits'] = df['ovisits'].astype('int8')
df['novisits'] = df['novisits'].astype('int8')
df['emergency'] = df['emergency'].astype('int8')
df['hospital'] = df['hospital'].astype('int8')
df['chronic'] = df['chronic'].astype('int8')
df['school'] = df['school'].astype('int8')
```
